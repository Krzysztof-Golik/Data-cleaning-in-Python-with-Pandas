```python
def format_phone_number(x):
    if pd.isna(x):
        return x  # Return NaN as is
    x = str(x).replace('nan', '').replace('Na', '')  # Remove unwanted strings
    x = x.strip()  # Remove leading/trailing whitespace
    if len(x) >= 7:  # Ensure the number is at least 7 digits long
        return f"{x[0:3]}-{x[3:6]}-{x[6:10]}"
    elif len(x) >= 3:  # If it's less than 7 but at least 3 digits, do a partial formatting
        return f"{x[0:3]}-{x[3:6]}"
    else:
        return x  # If it's less than 3 digits, return as is

# Apply the function to the Phone_Number column
df["Phone_Number"] = df["Phone_Number"].apply(format_phone_number)

print(df)
```
